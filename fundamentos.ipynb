### Explicacion

Variables that are created outside of a function (as in all of the examples above) are known as global variables.

Global variables can be used by everyone, both inside of functions and outside.

```python
x = "awesome"

def myfunc():
  print("Python is " + x)

myfunc()
```

> Las variables locales no pueden accederse fuera de la función.

```python
# Create a variable inside a function, with the same name as the global variable
x = "awesome"

def myfunc():
  x = "Purete"
  print("Python is " + x)

myfunc()
print(x)
```

### Keyword global

```python
# If you use the global keyword, the variable belongs to the global scope:
def mi_funcion():
    global msj 
    msj = "Tu mama"
    print(msj)

mi_funcion()
print(msj)
```

### Crear una clase e instanciar un objeto

Con *Class* creamos un disenho.

Que significa instanciar? 
*Significa:* crear en memoria un ejemplar de un conjunto de datos y código definido por una clase

```python
# Create a class named MyClass, with a property named x:

class MyClass:
  x = 5

# Now we can use the class named MyClass to create objects
# Create an object named p1, and print the value of x:

p1 = MyClass() # Se instancia el objeto
print(p1.x)
```

### The __init__() function

funcion built in.
Todas las clases tienen una funcion __init__() Es una funcion *constructora*, las funciones constructoras son las que inicializan (asignar valores) al objeto de la clase cuando es creada.

```python
# A Sample class with init method
class Person:

	# init method or constructor
	def __init__(self, name):
		self.name = name

	# Sample Method
	def say_hi(self):
		print('Hello, my name is', self.name)

# Instancio el objeto p
p = Person('Nikhil')
p.say_hi()
```

<center>
Nota: El parámetro self es una referencia a la instancia actual de la clase, y se utiliza para acceder a variables que pertenecen a la clase.
</center>
